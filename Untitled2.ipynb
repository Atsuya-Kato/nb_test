```python
#!/usr/local/bin/python

import os, sys
import json
import numpy as np
import matplotlib.pyplot as plt
import chainer
from chainer import cuda
from matplotlib import animation
from optparse import OptionParser

from util.cmap_bipolar import bipolar
import elecpy2 as elp

from matplotlib import animation, rc
from IPython.display import HTML

import time

%matplotlib inline
```

```python
# load sample parameter
with open (os.path.join('./temp_400*400/sim_params2.json'),'r') as f:
    sim_params = json.load(f)

# modify parameters
sim_params['time']['end'] = 500
sim_params["stimulation"]["extracellular"][0]["name"] = 'point'
sim_params["stimulation"]["extracellular"][0]["size"] = [100, 40, 5]
sim_params["stimulation"]["extracellular"][0]["start"] = 0.
sim_params["stimulation"]["extracellular"][0]["train"] = 4
sim_params["stimulation"]["extracellular"][0]['interval'] = 120
sim_params["stimulation"]["extracellular"][0]["duration"] = 10
sim_params["stimulation"]["extracellular"][0]["amplitude"] = 50
sim_params['log']['path'] = '/mnt/Omer/Project/07.DeepElectrodeMapping/kato/20190806/'
print json.dumps(sim_params, indent=4)
```

```python
if not os.path.exists(sim_params['log']['path']):
    os.makedirs(sim_params['log']['path'])
```

```python
with open(os.path.join(sim_params['log']['path'], 'sim_params.json'), 'w') as outfile:
    json.dump(sim_params, outfile)
```

```python
g = elp.sim_generator(sim_params)
```

```python
out = []
while True:
    img = g.next()
    if img is False: break
    out.append(np.copy(img))
```

```python
fig = plt.figure()
plt.axis('off')

ims = []
for img in out:
    im = plt.imshow(
        img.reshape(400,400),
        vmin = -100.0, vmax = 20.0,
        cmap=bipolar(neutral=0, lutsize=1024),
        interpolation='nearest')
    ims.append([im])

ani = animation.ArtistAnimation(fig, ims,interval =10)
#ani.save('anim.gif', writer="imagemagick")
#ani.save('anim.mp4', writer="ffmpeg")
#plt.show()

HTML(ani.to_html5_video())
```
