```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from elecpy.util.cmap_bipolar import bipolar
import h5py
from IPython.display import HTML
```

```python
ims=[]
fig = plt.figure()
input_file = "/mnt/Share/Temp/kato/sim_results/20200309_3/out.h5"
with h5py.File(input_file,"r") as f:
    for i in range(100):
        vmem= f['{:0=4}'.format(i)]['vmem'].value
        im = plt.imshow(vmem, vmin=-100, vmax=20,
                         cmap=bipolar(neutral=0, lutsize=1024),
                          interpolation='nearest')
        ims.append([im])
    
ani = animation.ArtistAnimation(fig, ims,interval=50)


HTML(ani.to_html5_video())
```
