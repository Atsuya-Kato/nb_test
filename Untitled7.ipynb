```python
import h5py
import numpy as np
```

```python
input_file = "./sim_results/20200130_2/out.h5"
with h5py.File(input_file,"r") as f:
    print('(1)----- 1st level -----')
    # ファイルオブジェクトをイテレートするとファイル直下のオブジェクト名を返す
    for k in f:
        print(k)
    print('(2)----- 2nd level -----')
    for k in f['0300']:
        print(k)
    print('(3)----- 3rd level -----')
    for k in f['0300/phie']:
        print(k)  
```

```python
x = np.ones((400,400))
x =np.where(x == 1, 1.41128048e-03 , 100)
```
